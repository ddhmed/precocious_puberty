```python
microbiome = pd.read_csv('../data/Preprocess/rawdata_16S_abundance.tsv', sep='\t', index_col=0)
microbiome.head()
```

```python
metabolism = pd.read_csv('../data/Preprocess/rawdata_metabolism_abundance.tsv', sep='\t', index_col=0)
metabolism.head()
```

```python
microbiome.shape, metabolism.shape
```

```python
overlap = set(microbiome.columns)&set(metabolism.columns)
overlap_normal = [i for i in overlap if 'N' in i]
overlap_PP = [i for i in overlap if 'PP' in i]
len(overlap), len(overlap_normal), len(overlap_PP)
```

```python
age_n = meta.loc[overlap_normal, 'Age']
age_n.mean(), 1.96 * age_n.std() / np.sqrt(len(age_n)) #95置信区间
```

```python
age_pp = meta.loc[overlap_PP, 'Age']
age_pp.mean(), 1.96 * age_pp.std() / np.sqrt(len(age_pp)) #95置信区间
```

```python
from scipy import stats
stats.ttest_ind(age_n,age_pp)
```
